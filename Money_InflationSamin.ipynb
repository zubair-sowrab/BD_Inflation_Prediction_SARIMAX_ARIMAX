```python
newDF= DF.drop(columns =['Interest_Rate', 'Inflation_Rate_MoM' ] ,axis=1)
```

```python
newDF.corr()
```

```python
sns.heatmap(newDF.corr(), annot=True, vmin=-1, vmax=1)
```

```python
X= newDF.iloc[:,:-1]
Y= newDF.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, random_state= 1)
```

#DecisionTreeRegression


```python
from sklearn.tree import DecisionTreeRegressor
dt= DecisionTreeRegressor()
dt.fit(X_train,Y_train)
```

```python
y1_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
accuracy = r2_score(Y_test, y1_pred)
print(accuracy)
```

#RandomForestRegression

```python
from sklearn.ensemble import RandomForestRegressor
rfr= RandomForestRegressor()
rfr.fit(X_train,Y_train)
```

```python
y2_pred= rfr.predict(X_test)
```

```python
accuracy02= r2_score(Y_test,y2_pred)
print(accuracy02)
```

```python
from sklearn.metrics import mean_absolute_error as mae
mae(Y_test,y2_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test,y2_pred, multioutput='raw_values')
```

```python
from sklearn.metrics import  mean_squared_log_error
mean_squared_log_error(Y_test,y2_pred)
```
